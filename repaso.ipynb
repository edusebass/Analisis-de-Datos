```python
a = open("dataprueba.txt", "r")
contenido = a.read()

palabras = contenido.split()
cantidad_palabras = len(palabras)
frecuenciaPalabras = {}

for palabra in palabras:
    if palabra in frecuenciaPalabras:
        frecuenciaPalabras[palabra] += 1
    else:
        frecuenciaPalabras[palabra] = 1
        
for palabra, frecuencia in frecuenciaPalabras.items():
    print(f'{palabra}: {frecuencia}')
a.close()

a = open("2.txt", "r")
contenido = a.read()

palabras = contenido.split()
cantidad_palabras1 = len(palabras)
frecuenciaPalabras = {}

for palabra in palabras:
    if palabra in frecuenciaPalabras:
        frecuenciaPalabras[palabra] += 1
    else:
        frecuenciaPalabras[palabra] = 1
        
for palabra, frecuencia in frecuenciaPalabras.items():
    print(f'{palabra}: {frecuencia}')
    
print(cantidad_palabras)
print(cantidad_palabras1)
print(cantidad_palabras1 + cantidad_palabras)
a.close()
```

```python
n = 12
X = np.arange(n)
Y1 = (1 - X / float(n)) * np.random.uniform(0.5, 1.0, n)
Y2 = (1 - X / float(n)) * np.random.uniform(0.5, 1.0, n)
plt.bar(X, +Y1, facecolor='#9999ff', edgecolor='white')
plt.bar(X, -Y2, facecolor='#ff9999', edgecolor='white')
for x, y in zip(X, Y1):
    plt.text(x + 0.4, y + 0.05, '%.2f' % y, ha='center', va='bottom')
    plt.ylim(-1.25, +1.25)
```

```python
import requests
```

```python
from bs4 import BeautifulSoup
```

```python
import couchdb
```

```python
couch = couchdb.Server('http://admin:admin@localhost:5984')
```

```python
db = couch['archivos']
```

```python
url = "https://latierramedia.com/categoria/blog-de-la-tierra-media/"
response = requests.get(url)
```

```python
response
```

```python
#ANALIZAR EL HTML DE LA PAGINA
soup = BeautifulSoup(response.text, "html.parser")
```

```python
links = soup.find_all('h2')
data = []

for link in links:
    href = link.get('href')
    text = link.text
    edu = link.edu
    data.append({'href': href, 'text td': text, 'Nombre': 'edu'})
```

```python
# Guardar los datos en CouchDB
for item in data:
    db.save(item)
print("Enviado con exito")
```
