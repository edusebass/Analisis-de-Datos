```python
frutas_costa = ["platano", "coco", "piña"]
```

```python
frutas_sierra = ["mandarina", "guaba", "chirimoya"]
```

```python
index = ["a", "b", "c"]
```

```python
todas_las_frutas = {"costa": frutas_costa, "sierra":frutas_sierra}
```

```python
todas_las_frutas
```

```python
#creamos un dataframe
frutas_indexadas = pd.DataFrame(todas_las_frutas)
```

```python
frutas_indexadas
```

```python
##cambiamos las etiquetas
frutas_etiquetadas = pd.DataFrame(todas_las_frutas, index = index)
```

```python
frutas_etiquetadas
```

```python
##crear archivos segun la extension
frutas_etiquetadas.to_csv("frutas.csv")
```

```python
fornite_df = pd.read_csv("fortnite.csv", index_col = 'index')
fornite_df
```

```python
fornite_df[ ((fornite_df['Placed'] < 4) & (fornite_df['Hits'] ))].head(3)
#Los 3 mejores jugadores con mayores hits
```

```python
fornite_df[ ((fornite_df['Eliminations'] > 4) & (fornite_df['Damage to Players'] > 400 ))].head(3)
#Las eliminaciones con mas daño al jugador
```

```python
fornite_df[ ((fornite_df['Placed'] <= 10) & (fornite_df['Mental State'] == "high"))].head()
#Los 5 primeros registros encontrados de jugadores con el estado mental elevado y consiguieron posicionarse dentro del top 10
```

```python
fornite_df[ ((fornite_df['Accuracy'] >= 0.30) & (fornite_df['Eliminations'] >= 5))].head()
#Los jugadores con medior presicion y con 5 o mas eliminaciones
```

```python
fornite_df[ ((fornite_df['Distance Traveled'] >= 3500 ) & (fornite_df['Materials Used'] > 500 ))].head()
#Los jugadores con mayor distancia recorrida y que mas de 500 de materiales
```

```python
fornite_df[ ((fornite_df['Damage to Structures'] >= 10000 ) & (fornite_df['Damage Taken'] > 200 ))].head()
#Los jugadores que hicieron mas daño a las estructuras pero recibieron mas daño
```
